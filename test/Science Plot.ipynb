```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use(['science','ieee'])
```

```python
counts1=pd.read_csv('H:\\0-毕业设计\\godetector\\test\\1node\\counts.csv')
counts2=pd.read_csv('H:\\0-毕业设计\\godetector\\test\\2node\\counts.csv')
counts3=pd.read_csv('H:\\0-毕业设计\\godetector\\test\\3node\\counts.csv')
```

```python
counts3.head()
```

```python
x1=[]
x2=[]
x3=[]
precount1=0
precount2=0
precount3=0
```

```python
for i in range(len(counts1)):
    count=counts1.iloc[i,1]
    for j in range(count-precount1):
        if counts1.iloc[i,0] > 275:
            continue
        x1.append(counts1.iloc[i,0])
    precount1=count
for i in range(len(counts2)):
    count=counts2.iloc[i,1]
    for j in range(count-precount2):
        if counts2.iloc[i,0] > 275:
            continue
        x2.append(counts2.iloc[i,0])
    precount2=count
for i in range(len(counts3)):
    count=counts3.iloc[i,1]
    for j in range(count-precount3):
        if counts3.iloc[i,0] > 275:
            continue
        x3.append(counts3.iloc[i,0])
    precount3=count
```

```python
norm1 = np.random.normal(0, 20, len(x1))
x1res = np.array(x1)+norm1
norm2 = np.random.normal(0, 20, len(x2))
x2res = np.array(x2)+norm2
norm3 = np.random.normal(0, 20, len(x3))
x3res = np.array(x3)+norm3
```

```python
plt.hist(x3, 50, density=False, color='b',alpha=0.5,label='Group1')
plt.hist(x2, 50, density=False, color='g',alpha=0.4,label='Group2')
plt.hist(x1, 50, density=False, color='r',alpha=0.3,label='Group3')

plt.xlabel('Latency/ms')
plt.ylabel('Count')
plt.title('Histogram of Latency') #,fontdict={'fontsize':8}
plt.legend(loc='upper right')
plt.savefig("Latency.png", dpi=2000)
plt.show()
```
